```python
def vacuumCleaner():
  #   define the goal state
  #   if state is 0 then it's clean, if 1 then state is dirty
  goal_state = {'A' : '0' , 'B' : '0'}
  
  #   set the cost variable initally to 0
  cost = 0

  #   take the start location of vacuum cleaner as the input
  start_location = str(input("Enter the start location of cleaner : "))
  
  #   condition of start state whether its clean or not
  status_input = input("Enter the status of the location "+ (start_location) + " : ")

  status_input_complement = input("Enter the status of other room : ")
  # print("Initial location condition "+str(goal_state))

  if start_location == 'A':
    print("The vacuum cleaner is at location A")

    if status_input == '1':
      #   location A is dirty suck the dirt and mark it as clean
      print("Location A is dirty")
      goal_state['A'] = '0'
      cost += 1   #   cost for sucking the dirt
      print("Cost for cleaning A "+str(cost))
      print("Location A has been cleaned")

      if status_input_complement == '1':
        #   location B is dirty
        print("Location B is dirty")
        print("Moving right to location B")
        #   moving right
        cost += 1
        #   suck the dirt in location B and compute cost
        goal_state['B'] = '0'
        cost += 1
        print("Cost for cleaning  "+str(cost))
        print("Location B has been cleaned")

      else:
        #   B is clean no action needed
        print("Cost for cleaning "+str(cost))
        print("Location B is already clean")
    
    else:
      # location A is already clean
      print("A is already clean")

      #   Move right to check the status of location A
      print("Moving right ")
      cost += 1
      
      if status_input_complement == '1':
        print("Location B is dirty ")
        cost += 1
        goal_state['B'] = 0
        print("All the blocks are cleaned")
        print("Total cost of cleaning is " + str(cost))
        
      else:
        print("All the states are clean")
        print("Cost of cleaning is 0")


  else:
    print("The vacuum cleaner is at position B")

    if status_input == '1':
      #location B is dirty
      print("Location B is dirty")

      cost += 1
      goal_state['B'] = '0'
      print("State B is cleaned")
      print("Cost of cleaning B is "+ str(cost))

      if status_input_complement == '1':
        # location A is dirty
        print("Location A is dirty")
        print("Moving left to location A")
        cost += 1
        goal_state['A'] = '0'
        cost += 1
        print("State A is cleaned")
        print("Total cost of cleaning is " + str(cost))
      
      else:
        #   location A is already in clean state
        print("Location A is already clean")
        print("Cost of cleaning is "+ str(cost))
      
    else:
      # location B is already clean
      print("B is already clean")

      #   Move left to check the status of location A
      print("Moving left ")
      cost += 1
      
      if status_input_complement == '1':
        print("Location A is dirty ")
        cost += 1
        goal_state['A'] = 0
        print("All the blocks are cleaned")
        print("Total cost of cleaning is " + str(cost))
        
      else:
        print("All the states are clean")
        print("Cost of cleaning is 0")
    
vacuumCleaner()
```
